## Reinforced Clostridial Medium (RCM)
http://www.oxoid.com/UK/blue/prod_detail/prod_detail.asp?pr=CM0149&org=53&c=UK&lang=EN

- Yeast Extract -> lots of stuff in it, unsure of what's garbage

    - Glutathione: gthrd (reduced version)
    - 18 AA's: for concentration see MediumMainComponents
    - Dextran: glc (it's made of glucose)
    - Mannan: mannan or mannose -> man
    - Trehalose: tre (not present in exchange reactions)
    - B-vitamins: (Clemence mentioned vitamins being important so I'm putting them all in)
        - Thiamin -> thm
        - Riboflavin -> ribflv
        - Niacin -> trp_L (made of tryptophan)
        - Pantothenic Acid -> pnto_R
        - Pyridoxine -> pydxn
        - Biotin -> btn
        - Folic Acid -> folate -> fol
        - Cobalamin -> b12 / cbl1
    - Biotin
- Peptone -> Source of AA. We stick to bacteriological peptone
- Glucose -> Assuming D_Glucose: glc__D
- Soluble Starch -> strch1 or strch2
- Sodium Chloride -> na1, cl
- Sodium Acetate -> ac, na1
- Cysteine Hydrochloride -> cys_L, cl, h (could also be cys_D? cys_L is more widely used)
- Agar: Agarose (70%) and Agaropectin (30%) according to Wiki
    - Agarose: D-galactose & 3,6-anhydro-L-galactopyranose
        - D-galactose: gal
        - 3,6-anhydro-L-galactopyranose: similar structure & compositon to beta-D-allose. D-Allose: all__D
    - Agaropectin: D-galactose, L-galactose, pyruvate, sulfate
        - D-galactose: gal
        - L-galactose: gal__L
        - Pyruvate: pyr (not present in exchange reactions)
        - Sulfate: so4
